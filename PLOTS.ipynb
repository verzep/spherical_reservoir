```python

#ATTENZIONE AGLI INDICI

def M(n, k, a):
    return (1 - 1/(a +1))**(k)#ATTENZIONE AGLI INDICI

def M_loss_1(n,k,t,a):
    return ((a *1./(a+1))**(n+t-k)) - (a*1./(a+1))**(n-k) 

def M_loss_2(A, D, a):
    return (a/(a+1))**A - (a/(a+1))**(A+D)



def M_l(n,k,a):
    return a**(k)#ATTENZIONE AGLI INDICI
    
def M_loss_1_l(n,k,t,a):
    return a**(t-1) * a**(n-k)
def M_loss_2_l(A,D,a):
    return (1-a)**D *a**A


def M_tanh(n,k,a):
    LISTA = []
    for k_ in range(k):
        x = np.tanh(1)
    for k_ in range(k):#ATTENZIONE AGLI INDICI
        for _ in range(k_):
            x = np.tanh(x*a)
        LISTA.append(x)
    return np.array(LISTA)
    
```

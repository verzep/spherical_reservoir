```python
e0 = np.load('E_test_0_bho.npy')
e5 = np.load('E_test_5.npy')
e7 = np.load('E_test_7.npy')

sr0 = np.load('SR_0_bho.npy')
sr5 = np.load('SR_5.npy')
sr7 = np.load('SR_7.npy')

scal0 = np.load('scal_0_bho.npy')
scal5 = np.load('scal_5.npy')
scal7 = np.load('scal_7 .npy')

e7[17][1] = np.load('E_test_7_part.npy')
sr7[17][1] = np.load('SR_7_part.npy')
scal7[17][1] = np.load('scal_7_part.npy')

L = 20
tau = np.arange(0,L)
nu = np.log(np.logspace(-2.5,1,L))
```
